```python
lastLine = ''
with open('movie.csv',encoding="utf-8") as file:
    for i,line in enumerate(file.readlines()):
        if "|" not in line:
            lastLine[3] += line.replace('\n', '')
        else: 
            lastLine = line.split(" | ")
            lastLine[3] = lastLine[3].replace('\n', '')
            with open('movieFinal.csv', 'a') as movieFile:
                movieFile.truncate(0)
                movieFile.write(str(lastLine))
```
